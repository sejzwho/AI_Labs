```python
i = (0,0)
res = (0,0)

def emptyA(res):
    a = res[1]
    print("Empty A:")
    return (0,a)

def emptyB(res):
    a = res[0]
    print("Empty B:")
    return (a,0)

def fullA(res):
    a = res[1]
    print("Full A:")
    return (A,a)

def fullB(res):
    a = res[0]
    print("Full B:")
    return (a,B)

def pourInA(res):
    a = res[0]
    b = res[1]
    temp = a+b
    if(temp > A):
        a = A
        b = temp-A
    if(temp <= A):
        a = temp
        b = 0
    print("Pour in A:")
    return (a,b)

def pourInB(res):
    a = res[1]
    print("Pour in B:")
    return (0,a)

x,y = input("Enter capacity of jug A and B").split()
A = int(x)
B = int(y)
current = (0,0)
gx,gy = input("Enter goal").split()
goal = (int(gx),int(gy))
prev = "None"
tup = True

while(tup):

    if(prev == "None"):
        current = fullB(current)
        print(current)
        prev = "fullB"
        if(current[0] == goal[0] and current[1] == goal[1]):
            break
    
    if(prev == "fullB"):
        current = pourInA(current)
        print(current)
        prev = "pourInA"
        if(current[0] == goal[0] and current[1] == goal[1]):
            break
        
    if(current[0] == A):
        current = emptyA(current)
        print(current)
        prev = "emptyA"
        if(current[0] == goal[0] and current[1] == goal[1]):
            break
        
    if(prev == "pourInA"):
        current = fullB(current)
        print(current)
        prev = "fullB"
        if(current[0] == goal[0] and current[1] == goal[1]):
            break
        
    if(prev == "emptyA"):
        current = pourInA(current)
        print(current)
        prev = "pourInA"
        if(current[0] == goal[0] and current[1] == goal[1]):
            break
        
    
        
```
